vou mostrar no exemplo abaixo a forma que querendo ou não seria mais pratica, mas que pode gerar mais dor de cabeça de acordo com a quantidade de dados.
VOU SELECIONAR AS INFORMAÇÕES DOS POSTOS DE SP E RJ QUE ESTEJAM COM A GASOLINA COMUM ACIMA DE 2 REAIS

```python
selecao_um = (info['ESTADO'] == 'RIO DE JANEIRO') | (info['ESTADO'] == 'SAO PAULO')
selecao_dois = (info['PRODUTO'] == 'GASOLINA COMUM')
selecao_treis = (info['PREÇO MÉDIO REVENDA'] > 2)

selecao_total = selecao_um & selecao_dois & selecao_treis
```

```python
selecao_total
```

```python
#escrevemos pouco mas tambem percoremos o dateframe todo para fazer comparações, podendo gerar lentidão 
info[selecao_total]
```

Agora vamos fazer de uma forma mais eficiênte, utilizando um pouco mais de variaveis, mas, ficando de uma forma mais leve

```python
selecao_um = (info['ESTADO'] == 'RIO DE JANEIRO') | (info['ESTADO'] == 'SAO PAULO')

postos_sp_rj = info[selecao_um]

#utilizaremos esse dateframe menor, assim sendo mais tranquilo de trabalhar...
postos_sp_rj
```

```python
selecao_dois = (postos_sp_rj['PRODUTO'] == 'GASOLINA COMUM')

postos_sp_rj_gasolina = postos_sp_rj[selecao_dois]

postos_sp_rj_gasolina
```

```python
selecao_treis = (postos_sp_rj_gasolina['PREÇO MÉDIO REVENDA'] > 2)
postos_sp_rj_maior_que_dois = postos_sp_rj_gasolina[selecao_treis]

postos_sp_rj_maior_que_dois
```

```python
info_pre['DATA INICIAL'] = pd.to_datetime(info_pre['DATA INICIAL'])
info_pre['DATA FINAL'] = pd.to_datetime(info_pre['DATA FINAL'])
```

```python
#convertendo atributos/colunas para numerico
for atributo in ['MARGEM MÉDIA REVENDA', 'PREÇO MÉDIO DISTRIBUIÇÃO', 'DESVIO PADRÃO DISTRIBUIÇÃO', 'PREÇO MÍNIMO DISTRIBUIÇÃO', 'PREÇO MÁXIMO DISTRIBUIÇÃO', 'COEF DE VARIAÇÃO DISTRIBUIÇÃO']:
#converte para 'numeric' e caso tenha valores que sejam identificados diferentemente de 'numeric' será aplicado no local
#um valor null ou NaN
    info_pre[atributo] = pd.to_numeric(info_pre[atributo], errors='coerce')
```
